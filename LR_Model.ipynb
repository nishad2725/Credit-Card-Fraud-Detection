# Under-Sampling 

## Build a sample dataset containing similar distribution of normal and fraudulent transactions .

```python
legit_sample = legit.sample(n=492)
```

Concatinating the two dataframs

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

### Splitting the data into features and targets

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
X.head()
```

```python
Y.head()
```

```python
Y
```

```python
Y.value_counts()
```

## split the data into training data and test data. 

```python
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model training 

```python
model = LogisticRegression()
```

```python
# Training the logistic Regression model with training data
```

```python
model.fit(X_train, Y_train)
```

## model evaluation 

```python
# Accuracy score on traiinig data

X_train_prediction = model.predict(X_train)
traininig_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('accuracy on training data: ',traininig_data_accuracy)
```

```python
# Accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('accuracy on training data: ',test_data_accuracy)
```
